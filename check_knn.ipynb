```python
import csv

with open('iris.data.txt', 'r') as csvfile:
    lines = csv.reader(csvfile)
    for row in lines :
        print (', '.join(row))

#split the data into a training dataset 

import random

def loadDataset(filename, split, trainingSet=[] , testSet=[]):
    with open(filename, 'r') as csvfile:
        lines= csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y] = float(dataset[x][y])
            if random.random() < split:
             #complete code
                trainingSet.append(dataset[x])
            else:
            #complete code
                testSet.append(dataset[x])

#test this function out with our iris dataset, as follows:

trainingSet=[]
testSet=[]
loadDataset('iris.data.txt', 0.66, trainingSet, testSet)
print ('Train: ' + repr(len(trainingSet)))
print ('Test: ' + repr(len(testSet)))
```

```python
import math

def euclideanDistance(instance1, instance2, length):
      #Complete the function
    dist=0
    for x in range(length):
        dist += pow((instance1[x]-instance2[x]),2)
    return math.sqrt(dist)

#test function
data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']

distance = euclideanDistance(data1, data2, 3)

print ('Distance: ' + repr(distance))
```

```python
import operator

def getNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
        distances.sort(key=operator.itemgetter(1))

    neighbors = []

    for x in range(k):
        neighbors.append(distances[x][0])

    return neighbors

#We can test out this function as follows:

trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]

testInstance = [5, 5, 5]

k = 1

neighbors = getNeighbors(trainSet, testInstance, 1)

print(neighbors)
```

```python
import operator

def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][ -1 ] #complete with appropriate number
        if response in classVotes:
        #Complete the if clause
            classVotes[response]+=1
        else:
            classVotes[response]=1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)

    return sortedVotes[0][0]

#We can test out this function with some test neighbors, as follows:

neighbors = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]

response = getResponse(neighbors)

print(response)
```

```python
def getAccuracy(testSet, predictions):
#Complete the function
    correct=0
    print("Length of testSet:", len(testSet))
    print("Length of predictions:", len(predictions))
    for x in range(len(testSet)):
        if testSet[x][-1]==predictions[x]:
            correct+=1
    return (correct/float(len(testSet))) * 100.0

#We can test this function with a test dataset and predictions, as follows:

testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]

predictions = ['a', 'a', 'a']

accuracy = getAccuracy(testSet, predictions)

print(accuracy)
```

```python
def main():
    trainingSet=[]
    testSet=[]
    loadDataset('iris.data.txt', 0.66, trainingSet, testSet)
    print ('Train: ' + repr(len(trainingSet)))
    print ('Test: ' + repr(len(testSet)))
    
    pred=[]
    k=2
    for x in range (len(testSet)):
        a=getNeighbors(trainingSet, testSet[x], k)
        b=getResponse(a)
        print(f'Predictions: {b}, Actuals: {testSet[x][-1]}')    
       
    c=getAccuracy(testSet, pred)
    print(f'Accuracy : {c}')
    
main()
```

```python
def manhattanDistance(instance1, instance2, length):
    dist = 0
    for x in range(length):
        dist += abs(instance1[x] - instance2[x])
    return dist

#test function
data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']

distance = manhattanDistance(data1, data2, 3)

print ('Distance: ' + repr(distance))
```
